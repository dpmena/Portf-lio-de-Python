```python
import pandas as pd
import matplotlib.pyplot as plt

# Criar um dataframe vazio para armazenar os gastos
gastos = pd.DataFrame(columns=["Categoria", "Valor"])

# Função para adicionar um gasto
def adicionar_gasto():
  categoria = input("Digite a categoria do gasto (ex: Alimentação, Transporte, Lazer): ").strip().title()
  try:
    valor = float(input("Digite o valor do gasto: "))
    if valor <= 0:
      print("O valor deve ser maior que zero. Tente novamente.")
      return
  except ValueError:
    print("Entrada inválida! Digite um número válido.")
    return

  global gastos
  novo_gasto = pd.DataFrame([[categoria, valor]], columns=["Categoria", "Valor"])
  gastos = pd.concat([gastos, novo_gasto], ignore_index=True)
  print("Gasto adicionado com sucesso!")

# Função para exibir o resumo dos gastos
def resumo_gastos():
  if gastos.empty:
    print("Nenhum gasto registrado ainda.")
    return
  print(("\nResumo dos gastos:"))
  print(gastos.groupby("Categoria").sum())
  print(f"Total gasto: R${gastos['Valor'].sum():.2f}")

# Função para gerar gráfico
def grafico_gastos():
  if gastos.empty:
    print("Nenhum gasto registrado para exibir o gráfico.")
    return
  resumo = gastos.groupby("Categoria").sum()
  resumo.plot(kind='bar', legend=False, figsize=(8,5), color='skyblue')
  plt.title("Gastos por Categoria")
  plt.xlabel("Categoria")
  plt.ylabel("Valor (R$)")
  plt.xticks(rotation=45)
  plt.grid(axis='y', linestyle='--', alpha=0.7)
  plt.show()

# Loop para adicionar gastos e exibir análise
enquanto = True
while enquanto:
  print("\n1 - Adicionar gasto")
  print("2 - Exibir Resumo")
  print("3 - Mostrar gráfico")
  print("4 - Sair")
  escolha = input("Escolha uma opção: ")

  if escolha == "1":
    adicionar_gasto()
  elif escolha == "2":
    resumo_gastos()
  elif escolha == "3":
    grafico_gastos()
  elif escolha == "4":
    print("Obrigada por usar o analisador de gastos!")
    enquanto = False
  else:
    print("Opção inválida! Escolha uma opção válida.")





```
